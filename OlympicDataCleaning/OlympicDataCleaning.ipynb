## Renaming

```python
new_names = {'Unnamed: 0': 'S.N',
            'Unnamed: 1': 'Country', 
           '? Summer': 'Summer Olympics',
           '01 !': 'Gold',
           '02 !': 'Silver',
           '03 !': 'Bronze',
           '? Winter': 'Winter Olympics',
           '01 !.1': 'Gold.1',
           '02 !.1': 'Silver.1',
           '03 !.1': 'Bronze.1',
           '? Games': 'Games', 
           '01 !.2': 'Gold.2',
           '02 !.2': 'Silver.2',
           '03 !.2': 'Bronze.2'}

df_olympics.rename(columns=new_names, inplace=1)
df_olympics.head()
```
